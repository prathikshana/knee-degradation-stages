### DATA EXTRACTION

```python
filepaths=[]  
labels=[]     #def class label
max_files=900 # set maximum number of files in a class as 900
data_dir=r'G:\knee' # floder where data resides
data_sep=['train', 'test', 'val'] # data sepration for algorithm evaluation 
class_totals=[0,0,0,0,0] # no of target classes

for s in data_sep:
    add_path=os.path.join(data_dir,s)  #joining data path
    classes=os.listdir(add_path) # list of classes    
    for i, iclass in enumerate(classes):  #itertate classes        
        clasadd_path=os.path.join(add_path, iclass)
        filelist=os.listdir(clasadd_path)
        for f in filelist:
            fpath=os.path.join(clasadd_path,f)
            # checks if the file extension is one that can be used with ImageDataGenerator
            index=fpath.rfind('.')
            ext=fpath[index+1:].lower()
            if ext in ['jpg', 'jpeg', 'png']: # check to see if the image files are in readable images
                try:
                    img = Image.open(fpath)  # open the image
                    img.verify()             # verify image format
                    if class_totals[i] < max_files:     # setting max file limit                                         
                        class_totals[i]=class_totals[i] + 1  #iterate till max limit reach
                        filepaths.append(fpath)             
                        labels.append(iclass)              
                except:
                    print ('file ', fpath, ' is an invalid image file and will not be processed') #if error in file
                
            else:
                print ('file ', fpath, ' with ', ext, '  will not be processed') #if error in file format
        
```

### DATA TRANSFORMATION

```python
# determine the class weight dictionary
for i in range(len(class_totals)): 
    print (' class: ', classes[i],  ' totals_samples: ', class_totals[i])  #Print total samples in each classes.
max_samples=np.max(class_totals)  

# Setting class weight
class_weight={}
for i in range (len(class_totals)):
    class_weight[i]=max_samples/class_totals[i]  
    
print(' total files: ',len(filepaths))  # print total files
```

### TRAIN TEST VAL SPLIT

```python
file_series=pd.Series(filepaths, name='filepaths') #make path
label_series=pd.Series(labels, name='labels')# make labels

df=pd.concat([file_series, label_series], axis=1) #merges files and label under df
train_split=.7 #train split 70
test_split=.15 # test 15

dummy_split=test_split/(1-train_split)

train_df, dummy_df=train_test_split(df, train_size=train_split, 
                                    shuffle=True, random_state=25)   #train data
test_df, valid_df=train_test_split(dummy_df, train_size=dummy_split,
                                   shuffle=True, random_state=25) #test data

print('train_df: ', len(train_df), '  test_df: ', len(test_df), ' valid_df: ', len (valid_df))
```

```python
batch_size=60 # setting batch size

# determine test batch size and test steps such that test_batch_size X test_steps = number of test samples
length= len(test_df) 
test_batch_size=sorted([int(length/n) for n in range(1,length+1) if length % n ==0
                        and length/n<=batch_size],reverse=True)[0]  
test_steps=int(length/test_batch_size) 
print (' test batch size: ', test_batch_size, '  test steps: ', test_steps)
```

### IMAGE PREPROCESSING

```python
image_shape=(224,224)  #image height and width
class_mode='categorical'  #target variable type
class_names=['NORMAL', 'TENTATIVE', 'MINIMAL', 'CONSERVATIVE', 'RELENTLESS'] #target varaiable class names

# create train, test, valid generators
def scaler(img):        # normal preprocessing function used originally but poor accuracy on test set
    return img/127.5 -1 # scale the pixels between -1 and + 1

def scalar(img):        #  adaptive thresholding on images to improve training accuracy
    img=np.array(img, dtype='uint8')
    img=cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)    
    th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, 
                                cv2.THRESH_BINARY, 21, 2);
    img=cv2.cvtColor(th2, cv2.COLOR_GRAY2RGB)
    img=np.array(img, dtype=('float32'))  
    img=img/127.5-1
    return img

#Image generator value defining
img_pre = tf.keras.preprocessing.image.ImageDataGenerator(preprocessing_function=scalar,
                                                          rotation_range= 15, width_shift_range=0.1,
                                                          height_shift_range=0.1, shear_range=0.2,
                                                        zoom_range=0.2,horizontal_flip=True )

gen= tf.keras.preprocessing.image.ImageDataGenerator(preprocessing_function=scalar)


#train-test-validation data generation from from train-test-valid split data
train_gen =img_pre.flow_from_dataframe(train_df, x_col='filepaths', y_col='labels',
                                       target_size=image_shape,class_mode=class_mode, 
                                       batch_size=batch_size, shuffle=True, random_state= 25)
test_gen =img_pre.flow_from_dataframe(test_df, x_col='filepaths', y_col='labels',
                                      target_size=image_shape,class_mode=class_mode,
                                      batch_size=test_batch_size, shuffle=False)
valid_gen =img_pre.flow_from_dataframe(valid_df, x_col='filepaths', y_col='labels',
                                       target_size=image_shape, class_mode=class_mode, 
                                       batch_size=batch_size, shuffle=False)
```

```python
def show_image_samples(gen, class_names):
    class_dict=gen.class_indices
    new_dict={}
    # make a new dictionary with keys and values reversed
    for key, value in class_dict.items(): # dictionary is now {numeric class label: string of class_name}
        new_dict[value]=key        
    images,labels=next(gen) # get a sample batch from the generator   
    plt.figure(figsize=(15, 20))
    length=len(labels)
    
    if length<15:   #show maximum of 15 images
        r=length
    else:
        r=15
        
    for i in range(r):
        plt.subplot(8, 4, i + 1)
        image=(images[i]+1 )/2 # scale images between 0 and 1 becaue pre-processor set them between -1 and +1
        plt.imshow(image)
        index=np.argmax(labels[i])
        class_name=class_names[index]
        plt.title(class_name, color='red', fontsize=15)
    plt.show()
```

```python
show_image_samples(train_gen, class_names) #sample train images
```

```python
show_image_samples(valid_gen, class_names) #sample validation images
```

```python
show_image_samples(test_gen, class_names) #sample test images
```

### MODEL BUILDING

```python
from tensorflow.keras.applications.vgg16 import VGG16

vgg_model = VGG16(input_shape = (224, 224, 3), # Shape of our images
include_top = False, # Leave out the last fully connected layer
weights = 'imagenet')
print(vgg_model.summary())
```

### TUNING VGG16 MODEL

```python
for layer in vgg_model.layers:
    layer.trainable = True
```

```python
# Flatten the output layer to 1 dimension
x = layers.Flatten()(vgg_model.output)

# Add a fully connected layer with 512 hidden units and ReLU activation
x1 = layers.Dense(512, activation='relu')(x)

# Add a dropout rate of 0.5
x2 = layers.Dropout(0.2)(x1)

x3 = layers.Dense(512, activation='relu')(x2)

# Add a final softmax layer for classification
x4 = layers.Dense(5, activation='softmax')(x3)

# MODEL COMPILATION
vggmodel = tf.keras.models.Model(vgg_model.input, x4)

vggmodel.compile(optimizer = tf.keras.optimizers.RMSprop(lr=0.0001),
              loss = 'categorical_crossentropy',metrics = ['acc'])
```

```python
# visualizing vgg16 model layers
print(vggmodel.summary())
```

```python
#fit and accuracy of model
hist = vggmodel.fit(train_gen, validation_data = valid_gen, 
                    steps_per_epoch = 2, epochs = 2)  #25 epochs trained
hist
```

```python
#Plotting acuracy and loss
import matplotlib.pyplot as plt
plt.plot(hist.history["acc"])
plt.plot(hist.history['loss'])
plt.title("model accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.legend(["Accuracy"," Loss"])
plt.show()
```

```python
#Test data accurarcy
print("Test Accuracy: ",vggmodel.evaluate(test_gen)[1])
```
